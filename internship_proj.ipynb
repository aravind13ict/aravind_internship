# Label encoding

```python
from sklearn.preprocessing import LabelEncoder
cols = ["Occupation","Type_of_Loan","Credit_Mix","Payment_of_Min_Amount","Payment_Behaviour","Credit_Score"]
le = LabelEncoder()
for col in cols:
    df[col] = le.fit_transform(df[col])
```

```python
#correlation
plt.figure(figsize=(30,18))
sns.heatmap(df.corr(),annot=True)
plt.title("Correlation Matrix")
plt.show()
```

# Creating Training Model

```python
#spliting into dependant and independant

x = df.drop('Credit_Score',axis=1)
y = df['Credit_Score']
```

```python
#Standardizing the features
from sklearn.preprocessing import StandardScaler
std_scaler=StandardScaler()
X=std_scaler.fit_transform(x)
X=pd.DataFrame(x)
```

```python
#splitting dataset into train and test   train=20%

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20,random_state=42)
```

# Logistic regression

```python
from sklearn import linear_model
lr=linear_model.LogisticRegression()
lr_model=lr.fit(x_train,y_train)
```

```python
lr_pred=lr_model.predict(x_test)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score,precision_score,recall_score,f1_score
print(accuracy_score(y_test, lr_pred))
```

```python
confusion_matrix(y_test,lr_pred)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier

metric_k = []
neighbors = np.arange(3, 20)
```

```python
for k in neighbors:
    classifier = KNeighborsClassifier(n_neighbors=k , metric = 'euclidean')
    classifier.fit(x_train, y_train)
    y_pred = classifier.predict(x_test)
    acc = accuracy_score(y_test, y_pred)
    metric_k.append(acc)
```

```python
metric_k
```

```python
#plotting the graph to find the k value

plt.plot(neighbors,metric_k,'o-')
plt.xlabel('k')
plt.ylabel('accuracy')
plt.show()
```

```python
#here lets take k as 5 as it gives the highest accuracy(values 4,5,6 has highest accuracy from that i take k as 5)
knn=KNeighborsClassifier(n_neighbors=5)
knn.fit(x_train,y_train)
y_pred_knn=knn.predict(x_test)
```

```python
print('Accuracy= ',accuracy_score(y_test,y_pred_knn))
```

```python
confusion_matrix(y_test,y_pred_knn)
```
