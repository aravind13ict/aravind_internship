since the skewness value of some of thed columns is less than 1/-1 , they have almost to normal distribution

Total_EMI_per_month,Amount_invested_monthly,Monthly_Balance these may have larger influence in dataset as a outlier

```python

Q1 = np.percentile(df['Total_EMI_per_month'],25,method = 'midpoint')
Q2 = np.percentile(df['Total_EMI_per_month'],50,method = 'midpoint')
Q3 = np.percentile(df['Total_EMI_per_month'],75,method = 'midpoint')

print("q1 is",Q1)
print("q2 is",Q2)
print("q3 is",Q3)

IQR = Q3-Q1
low_lim = Q1 - 1.5*IQR
up_lim = Q3 + 1.5*IQR

print('IQR is',IQR)
print('low_lim is',low_lim)
print('up_lim is',up_lim)
```

```python
#catching the outliers in Total_EMI_per_month and removing them
outlier = []
for x in df ["Total_EMI_per_month"]:
    if((x>up_lim)or(x<low_lim)):
        outlier.append(x)
```

```python
removed = df["Total_EMI_per_month"]>up_lim
df.loc[removed].index
df.drop(df.loc[removed].index,inplace = True)
```
